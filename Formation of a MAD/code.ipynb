## Figure S6

```python
g=6.673e-8
c=2.998e10
sunmass=1.989e33

bhmass=8.5
#field strength at the BH horizon 
b_h=0.7e8;

pi=2.0*math.asin(1.0)

r_g=g*sunmass*bhmass/c**2

bhspin=np.arange(1e-3, 0.99, 0.01)
r_h=r_g * (1.0+(1.0-bhspin**2)**0.5)
omeg_h=(c/2.0) * bhspin / r_h
p_bz=(b_h**2/4.0) * r_h**3 * omeg_h

fig=plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.semilogy(bhspin,p_bz,'k',linewidth=1.2);
ax.set_xlim(0.0, 1.0);
ax.set_ylim(1.0e35,1e38);
ax.tick_params(axis='both',direction='in',which='both',top=True, bottom=True, left=True, right=True, width=1.5)
ax.tick_params(which='major',length=8)
ax.tick_params(which='minor',length=3)
ax.spines['bottom'].set_linewidth(2);
ax.spines['left'].set_linewidth(2);
ax.spines['right'].set_linewidth(2);
ax.spines['top'].set_linewidth(2)
ax.set_xlabel("${\it a}$", fontsize=18);
ax.set_ylabel("${\it P}_{\mathrm {BZ}}\mathrm{ \ {(erg\; s^{-1})} }$", fontsize=18);
plt.xticks(fontsize=15);
plt.yticks(fontsize=15);

ax.xaxis.set_minor_locator(AutoMinorLocator(n=5))

fig.savefig("S6.png",dpi=600,bbox_inches='tight')
fig.savefig("S6.pdf",bbox_inches='tight')
```
